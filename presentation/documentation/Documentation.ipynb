# Solar System Simulation

## Overview

The Solar System Simulation is a Python-based project that visually represents the orbits of planets around the Sun. Utilizing the `pygame` library, the simulation provides an interactive graphical display of planetary motion, allowing users to observe the dynamics of our solar system.

## Project Components

1.  **`main.py`**:
    
    -   **Purpose**: Core script for the simulation.
    -   **Functionality**: Initializes the simulation, creates planet objects, and manages the main loop for updating and rendering planetary positions.
    -   **Key Classes**:
        -   `Planet`: Manages individual planets' properties and rendering.
        -   `Simulation`: Manages the overall simulation, including the display and event handling.
2.  **`constants.py`**:
    
    -   **Purpose**: Defines constants used throughout the project.
    -   **Contents**: Screen dimensions, colors, and planetary properties.
    -   **Usage**: Provides centralized configuration for screen size, colors, and planet attributes, ensuring consistency across the simulation.
3.  **`solarsystem_sim.py`**:
    
    -   **Purpose**: Contains additional or alternative simulation logic for rendering planetary orbits.
    -   **Functionality**: Provides a different approach or extended features compared to `main.py`, including potentially different settings or enhancements in the simulation.
    -   **Key Components**:
        -   **Initialization**: Sets up graphical rendering, initializes planets, and defines the simulation environment.
        -   **Planet Class**: Handles attributes and behaviors specific to planets, including position updates and drawing routines.
        -   **Simulation Logic**: Manages the main simulation loop, updating positions and rendering planets.

## Code Structure

The code is organized into functions and classes that handle different aspects of the simulation. Here’s an overview of the main components:

### `main.py`

The script is structured as follows:

1.  **Imports**:
    
    -   Imports necessary libraries, primarily `pygame` for graphical rendering.
2.  **Constants**:
    
    -   Defines constants for screen dimensions, colors, and other configuration settings.
3.  **Classes**:
    
    -   **`Planet`**:
        -   **Attributes**: `name`, `color`, `radius`, `distance`, `speed`, `angle`, `x`, `y`.
        -   **Methods**:
            -   `__init__()`: Initializes the planet with its properties.
            -   `update_position()`: Calculates and updates the planet's position based on its orbit.
            -   `draw()`: Draws the planet on the screen.
    -   **`Simulation`**:
        -   **Attributes**: Manages planets and the simulation state.
        -   **Methods**:
            -   `__init__()`: Sets up the display, initializes planets, and prepares the simulation.
            -   `run()`: Main loop that handles events, updates planet positions, and redraws the screen.
4.  **Main Function**:
    
    -   Initializes `pygame`, sets up the simulation, and starts the main loop.


### `constants.py`

The primary purpose of `constants.py` is to centralize and manage the configuration parameters for the simulation. By defining these values in one place, the code becomes easier to manage, and changes to configuration settings can be made efficiently.

The file typically contains the following types of constants:

1.  **Screen Dimensions**:
    
    -   `WIDTH`: The width of the simulation window.
    -   `HEIGHT`: The height of the simulation window.
2.  **Colors**:
    
    -   **General Colors**: Define colors using RGB tuples for various elements in the simulation.
3.  **Planetary Constants**:
    
    -   Constants related to the properties of planets, such as their size, colors, and distances from the Sun.

### `solarsystem_sim.py`


-   **Imports**: Imports necessary libraries, including `pygame` for rendering graphics.
    
-   **Constants**: Defines constants for screen dimensions, colors, and other configuration parameters.
    
-   **Classes**:
    
    -   **`Planet`**: Represents a planet with attributes such as size, color, orbital radius, and speed. Includes methods for updating the position and drawing the planet.
    -   **`Simulation`**: Manages the overall simulation, including initializing the display, handling user input, updating planetary positions, and rendering the simulation.
-   **Main Function**:
    
    -   Initializes `pygame` and sets up the display window.
    -   Creates instances of the `Planet` class for the Sun and various planets.
    -   Enters the main loop to continuously update the simulation state and redraw the screen.
